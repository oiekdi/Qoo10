```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pyperclip
import time

# ✅ 네이버 로그인 함수
def naver_login(driver, id, pw):
    print("1. 네이버 로그인 페이지로 이동합니다.")
    driver.get("https://nid.naver.com/nidlogin.login")
    time.sleep(1)

    print("2. 아이디 입력")
    pyperclip.copy(id)
    driver.find_element(By.CSS_SELECTOR, "#id").send_keys(Keys.CONTROL + "v")
    time.sleep(1)

    print("3. 비밀번호 입력")
    pyperclip.copy(pw)
    driver.find_element(By.CSS_SELECTOR, "#pw").send_keys(Keys.CONTROL + "v")
    time.sleep(1)

    print("4. 로그인 버튼 클릭")
    driver.find_element(By.CSS_SELECTOR, "#log\\.login").click()
    time.sleep(2)

    try:
        print("5. '등록안함' 버튼 클릭 시도")
        driver.find_element(By.CSS_SELECTOR, "#new\\.dontsave").click()
    except:
        print("5. '등록안함' 버튼 없음 → 넘어감")

    print("✅ 로그인 완료")

# ✅ 피드에서 URL 수집 함수
def get_urls(driver):
    print("6. 피드 페이지로 이동합니다.")
    driver.get("https://m.blog.naver.com/FeedList.naver")
    time.sleep(3)

    print("7. 게시글 링크 추출")
    elements = driver.find_elements(By.CSS_SELECTOR, "div[class^='text_area__'] > a[class^='link__']")
    urls = [el.get_attribute("href") for el in elements]

    print(f"✅ 총 {len(urls)}개의 URL 수집 완료")
    return urls

# ✅ 실행 흐름
if __name__ == "__main__":
    # 브라우저 옵션 설정
    options = Options()
    options.add_argument("--start-maximized")  # 창 최대화

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    # 네이버 로그인
    naver_login(driver, "vscria", "!fldifldigkwk1")

    # 게시글 URL 수집
    urls = get_urls(driver)

    print("\n📌 수집된 URL 목록:")
    for url in urls:
        print(url)

# 블로그로 이동
driver.get(url)
time.sleep(3)

# 제목 가져오기 
try:
    title = driver.find_element(By.CSS_SELECTOR, ".se-title-text").text
    print("제목:", title)
except Exception:
    print(f"{url} - 제목을 찾을 수 없습니다.")
    continue
print("제목:", title)

# 내용 가져오기
try:
    content = driver.find_element(By.CSS_SELECTOR, ".se-main-container").text
    print("내용:", content)
except Exception:
    print(f"{url} - 내용을 찾을 수 없습니다.")
    continue
print("내용:", content)

# 댓글 버튼 누르기
try:
    reply_button = driver.find_element(By.CSS_SELECTOR, "a.btn_reply")
    driver.execute_script("arguments[0].scrollIntoView(true);", reply_button)
    time.sleep(1)
    reply_button.click()
    time.sleep(3)
except Exception as e:
    print(f"{url} - 댓글 버튼 클릭 실패: {str(e)}")
    continue

   
# 댓글 창 활성화
try:
    driver.find_element(By.CSS_SELECTOR,".u_cbox_guide").click()
    time.sleep(1)
except Exception as e:
    print(f"{url} - 댓글 창 활성화 실패: {str(e)}")
    continue

# ai 댓글 생성
from google import genai
client = genai.Client(api_key="AIzaSyDaT2AcRn6NUaH-hDODh49sGmZoWY2RYL0")
response = client.models.generate_content(
    model="gemini-2.0-flash", contents=f"제목: {title}\n내용: {content} 위 제목과 내용을 분석해서 블로그를 잘 봤다며 제목과 내용에 어울리는 댓글을 작성해줘. 2줄 이내로 짧게 작성하고, 특수기호 및 이모지는 작성하지 마. 댓글 외에 다른 말은 쓰지 마."
)
comment = response.text

# 댓글 입력
try:
    driver.find_element(By.CSS_SELECTOR,"#naverComment__write_textarea").send_keys(comment)
    time.sleep(1)
    driver.find_element(By.CSS_SELECTOR,".__uis_naverComment_writeButton").click()
    time.sleep(3)
except Exception as e:
    print(f"{url} - 댓글 입력 실패: {str(e)}")
    continue
```
